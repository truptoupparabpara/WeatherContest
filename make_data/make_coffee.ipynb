```python
import numpy as np
import matplotlib as mlp
import pandas as pd
import datetime as dt
```

```python
bev_all = pd.read_csv("beverage_all.csv", encoding='utf8')
bev_juice = pd.read_csv("beverage_fruit.csv", encoding='utf8')
bev_latte = pd.read_csv("beverage_latte.csv", encoding='utf8')
bev_icetea = pd.read_csv("beverage_icetea.csv", encoding='utf8')
bev_coco = pd.read_csv("beverage_coco.csv", encoding='utf8')
bev_ia = pd.read_csv("beverage_ia.csv", encoding='utf8')
weather = pd.read_csv("weather_rmna.csv", encoding='cp949')
```

```python
weather = weather.rename(columns={"weather.ìŠ¤":"datetime"})
```

```python
bev_all['datetime']=bev_all['datetime'].apply(lambda e: pd.to_datetime(str(e), format="%Y-%m-%d"))
bev_juice['datetime']=bev_juice['datetime'].apply(lambda e: pd.to_datetime(str(e), format="%Y-%m-%d"))
bev_latte['datetime']=bev_latte['datetime'].apply(lambda e: pd.to_datetime(str(e), format="%Y-%m-%d"))
bev_icetea['datetime']=bev_icetea['datetime'].apply(lambda e: pd.to_datetime(str(e), format="%Y-%m-%d"))
bev_coco['datetime']=bev_coco['datetime'].apply(lambda e: pd.to_datetime(str(e), format="%Y-%m-%d"))
bev_ia['datetime']=bev_ia['datetime'].apply(lambda e: pd.to_datetime(str(e), format="%Y-%m-%d"))
```

```python
weather_mean = weather.groupby(['datetime'], as_index=False).mean()
```

```python
weather_mean['datetime'] = weather_mean['datetime'].apply(lambda e: pd.to_datetime(str(e), format="%Y-%m-%d"))
```

```python
weather_mean["rn_label"]=0
```

```python
weather_mean.loc[weather_mean["weather.sum_rn"]>10, "rn_label"]=2
weather_mean.loc[weather_mean["weather.sum_rn"]<10, "rn_label"]=1
weather_mean.loc[weather_mean["weather.sum_rn"]==0, "rn_label"]=0
```

```python
weather_mean["weekday"]=weather_mean["datetime"].dt.dayofweek
```

```python
weather_mean["season"]=0
```

```python
weather_mean.loc[weather_mean["datetime"].dt.month==1, "season"]=0
weather_mean.loc[weather_mean["datetime"].dt.month==2, "season"]=0
weather_mean.loc[weather_mean["datetime"].dt.month==3, "season"]=1
weather_mean.loc[weather_mean["datetime"].dt.month==4, "season"]=1
weather_mean.loc[weather_mean["datetime"].dt.month==5, "season"]=1
weather_mean.loc[weather_mean["datetime"].dt.month==6, "season"]=2
weather_mean.loc[weather_mean["datetime"].dt.month==7, "season"]=2
weather_mean.loc[weather_mean["datetime"].dt.month==8, "season"]=2
weather_mean.loc[weather_mean["datetime"].dt.month==9, "season"]=3
weather_mean.loc[weather_mean["datetime"].dt.month==10, "season"]=3
weather_mean.loc[weather_mean["datetime"].dt.month==11, "season"]=3
weather_mean.loc[weather_mean["datetime"].dt.month==12, "season"]=0
```

```python
weather_mean
```

```python
bev_all_weather = pd.merge(bev_all, weather_mean)
bev_juice_weather = pd.merge(bev_juice, weather_mean)
bev_latte_weather = pd.merge(bev_latte, weather_mean)
bev_coco_weather = pd.merge(bev_coco, weather_mean)
bev_ia_weather = pd.merge(bev_ia, weather_mean)
bev_icetea_weather = pd.merge(bev_icetea, weather_mean)
```

```python
bev_all_weather.to_csv("bev_all_weather.csv", mode='w', encoding='ms949')
bev_juice_weather.to_csv("bev_juice_weather.csv", mode='w', encoding='ms949')
bev_coco_weather.to_csv("bev_coco_weather.csv", mode='w', encoding='ms949')
bev_ia_weather.to_csv("bev_ia_weather.csv", mode='w', encoding='ms949')
bev_latte_weather.to_csv("bev_latte_weather.csv", mode='w', encoding='ms949')
bev_icetea_weather.to_csv("bev_icetea_weather.csv", mode='w', encoding='ms949')
```
