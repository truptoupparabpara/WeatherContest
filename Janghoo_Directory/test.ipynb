```python
import datetime
list_of_tweets = query_tweets('안녕', begindate=datetime.date(2019,1,1), enddate=datetime.date(2019,1,15), poolsize=150, lang='ko')
for tweet in list_of_tweets:
    print(tweet.fullname, tweet.timestamp, tweet.text)
    print("\n")
```
